```python
import zipfile
import pandas as pd
import os
import numpy as np
```

# Load multiple zip files, identifies csv and merge dataframes

```python
# create function that identifies all zip files in a location a return a list
def get_zip_files(directory_path):
    zip_files = []
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            if file.endswith(".zip"):
                zip_file_path = os.path.join(root, file)
                zip_files.append(zip_file_path)
    return zip_files
```

```python
# create a list with all the zip files of the survey chapters, exclude one
zip_list = get_zip_files(r'E:\pobreza_2021\data\emicron\2021')
zip_list.pop(5)
zip_list
```

```python
# create function to open zip files, open folders, identifies csv files, read it as dataframe and save it

def load_csv_from_zip(zip_file_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        for filename in zip_ref.namelist():
            if filename.endswith('.csv'):
                with zip_ref.open(filename) as csv_file:
                    df = pd.read_csv(csv_file, sep = ';')
                    df.columns = [col.lower() for col in df.columns]
                    return {zip_file_path: df}
    return None
```

```python
# use the previous function to iterate in all zip files, to store all dataframes in a dictionary
csv_dict = {}
for zip_file in zip_list:
    result = load_csv_from_zip(zip_file)
    if result is not None:
        csv_dict.update(result)
```

# Join all dataframe with a base dataframe selected

```python
# make a list with dictionary keys or dataframe names (exclude one of location)
zip_file_names = list(csv_dict.keys())
zip_file_names.pop(6)
zip_file_names
```

```python
# choose the base dataframe and iterate over the list to merge the other dataframes

base_df = csv_dict[zip_file_names[0]]

# jooin all the other dataframes to the base dataframe
for i in range(1, len(zip_file_names)):
    zip_file_name = zip_file_names[i]
    df = csv_dict[zip_file_name]
    df.drop(columns = ['f_exp', 'clase_te', 'cod_depto', 'area'], inplace = True)
    base_df = pd.merge(base_df, df, on = ['directorio', 'secuencia_p', 'secuencia_encuesta'], how = 'left', indicator='indicator_'+zip_file_name)
base_df.shape
```

```python
## make a function to create proportions of the variable of interest
def take_prop(var):
    df = base_df.groupby(var,group_keys=False)['f_exp'].sum().reset_index()
    df['proportions'] = ((df['f_exp']/df['f_exp'].sum())*100)
    df = df.sort_values(by = "proportions", ascending = False)
    return df
```
